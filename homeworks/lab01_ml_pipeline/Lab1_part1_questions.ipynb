## ex. 1

$$  
y = x^Tx,  \quad x \in \mathbb{R}^N 
$$

$$
\frac{dy}{dx} = \dfrac{d\left< x,x \right>}{dx} = \dfrac{\left< dx,x \right> + \left< x,dx \right>}{dx} = \dfrac{\left< 2x,dx \right>}{dx} = 2x
$$ 

## ex. 2

$$ y = tr(AB) \quad A,B \in \mathbb{R}^{N \times N} $$ 

$$
\frac{dy}{dA} =\dfrac{d\left<I, AB \right>_F}{dA} = \dfrac{\left<I, (dA)B \right>_F}{dA} = \dfrac{\left<B^T, dA\right>_F}{dA} = B^T
$$
